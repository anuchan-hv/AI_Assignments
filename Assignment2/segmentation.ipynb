```
import cv2
import os
import numpy as np
from sklearn.model_selection import train_test_split

IMAGE_WIDTH = IMAGE_HEIGHT = 384

def get_data(dir1, dir2):
    data = []
    for file in os.listdir(dir1):
        img = cv2.imread(dir1+file)
        data.append(img)

    for file in os.listdir(dir2):
        img = cv2.imread(dir2+file)
        data.append(img)

    return data

data = get_data('/content/melanoma_scaled/','/content/others_scaled/')
data = np.array(data)

labels = get_data('/content/melanoma_scaled_bin/','/content/others_scaled_bin/')
labels = np.array(labels)

print data.shape
print labels.shape
np.save('data.npy',data)
np.save('labels.npy',labels)

x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size = 0.2, random_state=42)
print x_train.shape, x_test.shape, y_train.shape, y_test.shape
np.save('train_data.npy',x_train)
np.save('test_data.npy',x_test)
np.save('train_labels.npy',y_train)
np.save('test_labels.npy',y_test)
```

```
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, Conv2DTranspose, UpSampling2D
import numpy as np
import keras.backend as K
from sklearn.model_selection import train_test_split

BATCH_SIZE = 32
EPOCHS = 10

def create_model():
    model = Sequential()
    
    model.add(Conv2D(20, kernel_size=(3, 3), activation='relu', input_shape=(384,384,3)))
    model.add(Conv2D(20, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

    model.add(Conv2D(30, (3, 3), activation='relu'))
    model.add(Conv2D(30, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(UpSampling2D(size=(2, 2)))
    model.add(Conv2DTranspose(30, kernel_size=(3, 3), activation='relu'))
    model.add(Conv2DTranspose(30, kernel_size=(3, 3), activation='relu'))

    model.add(UpSampling2D(size=(2, 2)))
    model.add(Conv2DTranspose(20, kernel_size=(3, 3), activation='relu'))
    model.add(Conv2DTranspose(3, kernel_size=(3, 3), activation='relu'))

    return model

def dice_coef(y_true, y_pred, smooth=1):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def dice_coef_loss(y_true, y_pred):
    return -dice_coef(y_true, y_pred)


train_data = np.load('train_data.npy')
train_labels = np.load('train_labels.npy')

train_data, val_data, train_labels, val_labels = train_test_split(train_data, train_labels, test_size = 0.2, random_state=42)

print train_data.shape, val_data.shape, train_labels.shape, val_labels.shape

model = create_model()
model.compile(loss=dice_coef_loss, optimizer='adam', metrics=['accuracy', dice_coef])
model.summary()
model.fit(train_data, train_labels, validation_data=(val_data,val_labels), batch_size=BATCH_SIZE, epochs=EPOCHS)

model.save('my_model.h5')
```

```
test_data = np.load('test_data.npy')
test_labels = np.load('test_labels.npy')

loss_and_metrics = model.evaluate(test_data, test_labels, batch_size=32, verbose=0)
print "Accuracy on test data: "+ str(100*loss_and_metrics[1])+"%"
```
