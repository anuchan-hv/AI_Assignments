```
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import os

datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

def augment(dir):
    for file in os.listdir(dir):
        img = load_img(dir+file)  
        x = img_to_array(img)  
        x = x.reshape((1,) + x.shape)  

        i = 0
        for batch in datagen.flow(x, batch_size=1,
                                  save_to_dir='melanoma_scaled_class1', save_prefix='ISIC', save_format='jpg'):
            i += 1
            if i > 4:
                break  

augment('melanoma_scaled_class1/')
```

```
import cv2
import numpy as np
import os
from sklearn.model_selection import train_test_split

def get_data(dir1,dir2):
    data = []
    for file in os.listdir(dir1):
        img = cv2.imread(dir1+file)
        data.append(img)

    for file in os.listdir(dir2):
        img = cv2.imread(dir2+file)
        data.append(img)

    return data

data = get_data('melanoma_scaled_class1/','others_scaled_class/')
data = np.array(data)

labels = np.ones(2074)
labels = np.append(labels, np.zeros(1626))
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size = 0.2, random_state=42)

print train_data.shape, test_data.shape, train_labels.shape, test_labels.shape

np.save('train_data.npy',train_data)
np.save('test_data.npy',test_data)
np.save('train_labels.npy',train_labels)
np.save('test_labels.npy',test_labels)
```

```
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.optimizers import adagrad, adam
from keras import optimizers
import os
from keras import utils
from sklearn.model_selection import train_test_split
import numpy as np
import keras.backend as K

IMG_WIDTH = IMG_HEIGHT = 384

batch_size = 8
num_classes = 2
epochs = 10
  
def recall(y_true, y_pred):
    """Recall metric.

    Only computes a batch-wise average of recall.

    Computes the recall, a metric for multi-label classification of
    how many relevant items are selected.
    """
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision(y_true, y_pred):
    """Precision metric.

    Only computes a batch-wise average of precision.

    Computes the precision, a metric for multi-label classification of
    how many selected items are relevant.
    """
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision
    
def f1(y_true, y_pred):
    p = precision(y_true, y_pred)
    r = recall(y_true, y_pred)
    return 2*((p*r)/(p+r+K.epsilon()))
  
train_data = np.load('train_data.npy')
train_labels = np.load('train_labels.npy')

train_data, val_data, train_labels, val_labels = train_test_split(train_data, train_labels, test_size = 0.2, random_state=42)

print train_data.shape, val_data.shape, train_labels.shape, val_labels.shape
  
model2 = Sequential()
model2.add(Conv2D(16, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)))
model2.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

model2.add(Conv2D(32, (3, 3), activation='relu'))
model2.add(Dropout(0.2))
model2.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

model2.add(Conv2D(64, (3, 3), activation='relu'))
model2.add(Dropout(0.2))
model2.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

model2.add(Flatten())
model2.add(Dropout(0.2))
model2.add(Dense(100))
model2.add(Activation('relu'))
model2.add(Dropout(0.2))
model2.add(Dense(50))
model2.add(Activation('relu'))
model2.add(Dropout(0.2))
model2.add(Dense(1))
model2.add(Activation('sigmoid'))

adam = optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
  
model2.compile(loss='binary_crossentropy',
          optimizer= adam,
          metrics=[f1, precision, recall])

model2.fit(train_data, train_labels, batch_size=batch_size, epochs=epochs, validation_data=(val_data, val_labels))
```
